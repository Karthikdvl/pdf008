```python
class NumberFinder:
    def __init__(self, arr):
        self.arr = arr

    def smallest_unique(self):
        # Flatten the 2D list
        flat = [num for sublist in self.arr for num in sublist]
        
        # Count occurrences of each number
        from collections import Counter
        counts = Counter(flat)
        
        # Filter only unique numbers (count == 1)
        unique_numbers = [num for num, count in counts.items() if count == 1]
        
        # Return smallest if exists
        if unique_numbers:
            return min(unique_numbers)
        else:
            return None  # No unique number

# Example usage
arr = [[1,2,2,3,3], [4,5,5,6,6], [7,8,8,9,9]]
obj = NumberFinder(arr)
print("Smallest unique number:", obj.smallest_unique())
```

```python
# searching and sorting 
# liner_search
def liner_search(arr, length, key):
    for i in range(0,length):
        if arr[i] == key:
            return i
        
    return -1

arr = [1,2,3,4,5,6,7]
length = len(arr)
key = int(input("enter a value to find"))

result = liner_search(arr,length, key)
if result == -1:
    print("element {} not found".format(key))
else:
    print("element {} found at index {}".format(key,result))
```

```python
# binary search
def binary_search(arr, left, right, key):
    if (right >=left):
        mid = (left + right) // 2  # Use integer division for the index
        if arr[mid] == key:
            return mid
        elif arr[mid] > key:
            return binary_search(arr, left, mid - 1, key)
        else:
            return binary_search(arr, mid + 1, right, key)
    else:
        return -1    

arr = [1,2,3,4,5,6,7]
length = len(arr)
key = int(input("enter a value to find"))
result = binary_search(arr, 0, length-1, key)
if result == -1:
    print("element {} not found".format(key))
else:
    print("element {} found at index {}".format(key,result))
```
