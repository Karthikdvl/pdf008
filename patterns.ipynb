```python
# pyramid pattern
class pyramid:
    def __init__(self, n):
        self.n = n
    def display(self):
        for i in range(1, self.n):
            for j in range(self.n -i -1):
                print(" ",end="")
                
            for k in range(1, 2*i):
                print("*", end="")
            print()
p = pyramid(5)
p.display() 
```

```python
#Pyramid Patterns in Python with Alphabet
n = 5
A = 65
for i in range(1,n):
    for j in range(n-i):
        print(" ", end="")
    for k in range(1,2*i):
        print(chr(A), end="")
        A += 1
    A = 65
    print()
        
```

```python
#Pyramid Patterns in Python with Alphabet
n = 5
A = 65
for i in range(1,n):
    for j in range(n-i):
        print(" ", end="")
    for k in range(1,2*i):
        print(chr(A), end="")
        A += 1
    A = 65
    for j in range(i-2,n):
        print(" ", end="")
    for k in range(1,2*i):
        print(chr(A), end="")
        A += 1
    A = 65
    print()
        
```
