```python
#binary to decimal
binary = input("enter a binary number")
decimal = 0
length = len(binary)
for i in range(length):
    bit = int(binary[length -i-1])
    decimal+= bit*(2**i)
print("the decimal value is",decimal)
```

```python
# counting vowels and consonants, special charter in string.
s = input("enter a stirng")
vowels = 'aeiouAEIOU'
vowel_count = 0
consonent_count = 0
special_char_count = 0
for c in s:
    if c.isalpha():
        if c in vowels:
            vowel_count+=1
        else:
            consonent_count+=1
    else:
        special_char_count+=1
print("the number of vowels are",vowel_count)
print("the number of consonents are",consonent_count)
print("the number of special characters are",special_char_count)
```

```python
#factorial using recursion
def factorial(n):
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
    
n = int(input("enter a number"))
fact = factorial(n)
print("the factorial of",n ,"is ", fact)

#fatorial using iteration
f = 1
if n==0 or n==1:
    f=1
else:
    for i in range(1,n+1):
        f =f*i
print("the factorial using iteration",f)
```
