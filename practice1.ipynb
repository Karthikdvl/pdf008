```python
class student:
    def __init__(abc,num1, num2):
        abc.num1 = num1
        abc.num2 = num2
    def average(abc):
            return (abc.num1 + abc.num2) / 2
    def multiplication(abc):
        return (abc.num1 * abc.num2)
        
s1 = student(80, 90)
print("Average:", s1.average())
s2 = student(5, 4)
print("Multiplication:", s2.multiplication())
```

```python
class text:
    def __init__(self, content):
        self.content = content
    def word_count(self):
        return len(self.content.split())
    def reversed(self):
        return self.content[::-1]
    def check_palindrome(self):
        if self.content == self.content[::-1]:
            return True
    
t1 = text("Hello World Python")
print("Word Count:", t1.word_count())
print("Reversed Content:", t1.reversed())
t2 = text("madam")
print("Is Palindrome:", t2.check_palindrome())
```

```python
#find the maximum element in a list,
# find n' th largest element in a list
class max_finder:
    def __init__(self, numbers):
        self.numbers = numbers
    def find_max(self):
        return max(self.numbers)
    def find_nth_largest(self, numbers, n):
        unique_numbers = list(set(numbers))
        unique_numbers.sort(reverse = True)
        if n<= len(unique_numbers):
            return unique_numbers[n-1]
        else:
            return None
mf = max_finder([3, 1, 4, 4, 5, 2])
print("Maximum Element:", mf.find_max())
print("2nd Largest Element:", mf.find_nth_largest([3, 1, 4, 4, 5, 2], 2))
```

```python
# find missing number in a list of consecutive numbers
class missing_number_finder:
    def __init__(self, numbers):
        self.numbers = numbers
    def find_missing_number(self, start, end):
        full_set = set(range(start, end+1))
        given_set = set(self.numbers)
        missing_number = full_set - given_set
        if missing_number:
            return missing_number.pop()
        else:
            return None
mnf = missing_number_finder([1, 2, 4, 5, 6])
print("Missing Number:", mnf.find_missing_number(1, 6))
```

```python
# find the intersection of two lists
class intersection_finder:
    def __init__(self, list1, list2):
        self.list1 = list1
        self.list2= list2
        def find_intersection(self):
            return list(set(self.list1) & set (self.list2))
        
if __name__ == "__main__":
    inf = intersection_finder([1, 2, 3, 4], [3, 4, 5, 6])
    print("Intersection:", inf.find_intersection())
    
```

```python
#count the occurrence of each element in a list
class occurrence_counter:
    def __init__(self, numbers):
        self.numbers = numbers
    def count_occurence(self):
        occurrence_dict = {}
        for num in self.numbers:
            if num in occurrence_dict:
                occurrence_dict[num] += 1
            else:
                occurrence_dict[num] = 1
        return occurrence_dict
oc = occurrence_counter([1, 2, 2, 3, 3, 3])
```

```python
def merge_sorted_lists(list1, list2):
    merged_list = []
    i = 0  # Pointer for list1
    j = 0  # Pointer for list2

    while i < len(list1) and j < len(list2):
        if list1[i] <= list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1

    # Add any remaining elements from list1
    while i < len(list1):
        merged_list.append(list1[i])
        i += 1

    # Add any remaining elements from list2
    while j < len(list2):
        merged_list.append(list2[j])
        j += 1

    return merged_list

# Example usage:
list_a = [1, 3, 5, 7]
list_b = [2, 4, 6, 8, 10]
result = merge_sorted_lists(list_a, list_b)
print(result)
```

```python
def longest_unique_substring_naive(s):
    res = ""
    for i in range(len(s)):
        seen = set()
        curr = ""
        for j in range(i, len(s)):
            if s[j] in seen:
                break
            seen.add(s[j])
            curr += s[j]
        if len(curr) > len(res):
            res = curr
    return res, len(res)
```

```python
A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

# Union (A or B)
union_set = A | B
print(f"Union: {union_set}") # Output: {1, 2, 3, 4, 5, 6}

# Intersection (A and B)
intersection_set = A & B
print(f"Intersection: {intersection_set}") # Output: {3, 4}

# Difference (A only)
difference_set = A - B
print(f"Difference: {difference_set}") # Output: {1, 2}
```
